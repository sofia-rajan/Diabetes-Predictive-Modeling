```python
    
# Method 1

# Method 1: Logistic Regression

import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Load the dataset
#url = "/Users/sivasankaranrajasekaran/Desktop/UTD/AML/Project/archive (5)/diabetes_binary_health_indicators_BRFSS2015.csv"
df = pd.read_csv("Diabetes_dataset.csv")

# Display the first few rows of the dataset
#print(df.head())

# Define features and target variable
X = df.drop("Diabetes_binary", axis=1)
y = df["Diabetes_binary"]

# Data Splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the Logistic Regression model
logreg = LogisticRegression()

# Hyperparameter tuning using GridSearchCV
param_grid_logreg = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
grid_logreg = GridSearchCV(logreg, param_grid_logreg, cv=5)
grid_logreg.fit(X_train, y_train)

y_pred_grid_test = grid_logreg.predict(X_test)

# Report coefficients and intercept
print("Coefficients:", grid_logreg.best_estimator_.coef_)
print("Intercept:", grid_logreg.best_estimator_.intercept_)

# Calculate and report the accuracy
print("Test Accuracy is:", grid_logreg.best_estimator_.score(X_test, y_test))

# Report the best hyperparameters
print("\nLogistic Regression Best Hyperparameters:", grid_logreg.best_params_)

# Get feature names
feature_names_logreg = X.columns

# Create a DataFrame to display feature names and their coefficients
coef_df_logreg = pd.DataFrame({"Feature": feature_names_logreg, "Coefficient": grid_logreg.best_estimator_.coef_[0]})
coef_df_logreg = coef_df_logreg.sort_values(by="Coefficient", ascending=False)

# Display the most important features
print("\nTop contributing features (Logistic Regression):")
print(coef_df_logreg)

###########################################################################################

# Method 2

# Method 2: Logistic Regression with Medical Conditions and Lifestyle Habits

# Categorize features into "Medical Conditions" and "Lifestyle Habits"
medical_conditions = ['HighBP', 'HighChol', 'Stroke', 'HeartDiseaseorAttack']
lifestyle_habits = ['BMI', 'Smoker', 'PhysActivity', 'Fruits', 'Veggies', 'HvyAlcoholConsump']

# Select features for each category
X_medical_conditions = X[medical_conditions]
X_lifestyle_habits = X[lifestyle_habits]

# Data Splitting for Medical Conditions
X_train_medical, X_test_medical, y_train, y_test = train_test_split(X_medical_conditions, y, test_size=0.3, random_state=42)

# Create and train the Logistic Regression model for Medical Conditions
logreg_medical = LogisticRegression()

# Hyperparameter tuning using GridSearchCV
param_grid_logreg_medical = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
grid_logreg_medical = GridSearchCV(logreg_medical, param_grid_logreg_medical, cv=5)
grid_logreg_medical.fit(X_train_medical, y_train)

y_pred_grid_test_medical = grid_logreg_medical.predict(X_test_medical)

# Report coefficients and intercept
print("Coefficients:", grid_logreg_medical.best_estimator_.coef_)
print("Intercept:", grid_logreg_medical.best_estimator_.intercept_)

# Calculate and report the accuracy
print("Test Accuracy is:", grid_logreg_medical.best_estimator_.score(X_test_medical, y_test))

# Report the best hyperparameters
print("\nLogistic Regression (Medical Conditions) Best Hyperparameters:", grid_logreg_medical.best_params_)

# Get feature names
feature_names_medical = X_medical_conditions.columns

# Create a DataFrame to display feature names and their coefficients for Medical Conditions
coef_df_medical = pd.DataFrame({"Feature": feature_names_medical, "Coefficient": grid_logreg_medical.best_estimator_.coef_[0]})
coef_df_medical = coef_df_medical.sort_values(by="Coefficient", ascending=False)

# Display the most important features for Medical Conditions
print("\nTop contributing features (Logistic Regression - Medical Conditions):")
print(coef_df_medical)

# Data Splitting for Lifestyle Habits
X_train_lifestyle, X_test_lifestyle, y_train, y_test = train_test_split(X_lifestyle_habits, y, test_size=0.3, random_state=42)

# Create and train the Logistic Regression model for Lifestyle Habits
logreg_lifestyle = LogisticRegression()

# Hyperparameter tuning using GridSearchCV
param_grid_logreg_lifestyle = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
grid_logreg_lifestyle = GridSearchCV(logreg_lifestyle, param_grid_logreg_lifestyle, cv=5)
grid_logreg_lifestyle.fit(X_train_lifestyle, y_train)

y_pred_grid_test_lifestyle = grid_logreg_lifestyle.predict(X_test_lifestyle)

# Report coefficients and intercept
print("Coefficients:", grid_logreg_lifestyle.best_estimator_.coef_)
print("Intercept:", grid_logreg_lifestyle.best_estimator_.intercept_)

# Calculate and report the accuracy
print("Test Accuracy is:", grid_logreg_lifestyle.best_estimator_.score(X_test_lifestyle, y_test))


# Report the best hyperparameters
print("\nLogistic Regression (Lifestyle Habits) Best Hyperparameters:", grid_logreg_lifestyle.best_params_)

# Get feature names
feature_names_lifestyle = X_lifestyle_habits.columns

# Create a DataFrame to display feature names and their coefficients for Lifestyle Habits
#coef_df_lifestyle = pd.DataFrame({"Feature": feature_names_lifestyle, "Coefficient": grid_logreg_lifestyle.best_estimator_.coef_[0]})
#coef_df_lifestyle = coef_df_lifestyle.sort_values(by="Coefficient", ascending=False)

# Display the most important features for Lifestyle Habits
print("\nTop contributing features (Logistic Regression - Lifestyle Habits):")
#print(coef_df_lifestyle)
```
